```python
import json
import numpy as np

from pathlib import Path
```

## The important paths

```python
dataset_path = Path("/work/mbonse/s4/30_data/EpsEri_0100_C-0656-A.hdf5")
```

```python
# other parameters
data_tag="13_object_final"
psf_template_tag="10_psf"
para_tag="header_13_object_final/PARANG"
```

## Create the config files

```python
config_path = Path("/work/mbonse/s4/70_results/01_closed_form_hyperparameters/configs_norm/")
save_path_root = Path("/work/mbonse/s4/70_results/01_closed_form_hyperparameters/trained_models_norm/")

lambda_regs = np.logspace(1, 8, 50)
mask_sizes = [4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0]
convolve = True
cut_radius_psf = 4.0
use_normalization = True
```

```python
config_path = Path("/fast/mbonse/s4/70_results/01_closed_form_hyperparameters/configs_no_norm/")
save_path_root = Path("/fast/mbonse/s4/70_results/01_closed_form_hyperparameters/trained_models_no_norm/")

lambda_regs = np.logspace(1, 8, 50)
mask_sizes = [4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0]
convolve = True
cut_radius_psf = 4.0
use_normalization = False
```

```python
idx = 0
for tmp_lambda_reg in lambda_regs:
    for tmp_mask_size in mask_sizes:
        
        tmp_save_name = "s4_model_lamb_{:.2f}".format(tmp_lambda_reg) + \
            "_mask_" + str(tmp_mask_size) + ".pkl"
        print(tmp_save_name)
        
        tmp_config = dict()
        tmp_config["dataset_path"] = str(dataset_path)
        tmp_config["data_tag"] = data_tag
        tmp_config["psf_template_tag"] = psf_template_tag
        tmp_config["para_tag"] = para_tag
        
        tmp_config["lambda_reg"] = tmp_lambda_reg
        tmp_config["mask_size"] = tmp_mask_size
        tmp_config["cut_radius_psf"] = cut_radius_psf
        tmp_config["convolve"] = convolve
        tmp_config["save_file"] = str(save_path_root / tmp_save_name)
        tmp_config["use_normalization"] = use_normalization
        
        save_config = str(config_path / ("config_" + str(idx).zfill(3) + ".json"))
        
        with open(save_config, 'w') as outfile:
            json.dump(tmp_config, outfile)
        
        
        idx += 1
```
