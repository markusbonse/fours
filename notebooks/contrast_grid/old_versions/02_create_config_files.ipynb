# Config 3 - each dataset with all limit setups

```python
all_lines = []

for tmp_dataset_name, lambda_reg in dataset_names:
    print(tmp_dataset_name)
    lines = []
    
    # 1. Load the pickle file
    with open("/fast/mbonse/s4/70_results/03_contrast_grid/pca_baseline/limits/" 
              + tmp_dataset_name + "/pca_limit_results.pkl", 'rb') as f:
        x = pickle.load(f)
    
    # 2. get the idx list
    setups = []

    for tmp_result in x.values():
        for tmp_mag_result in tmp_result:
            setups.append(tmp_mag_result[1])
            
    print(setups)
    
    # create the lines
    for tmp_setup in setups:
        lines.append(tmp_dataset_name + " " + str(tmp_setup).zfill(4) + "a " + str(lambda_reg))
        lines.append(tmp_dataset_name + " " + str(tmp_setup).zfill(4) + "b " + str(lambda_reg))
        lines.append(tmp_dataset_name + " " + str(tmp_setup).zfill(4) + "c " + str(lambda_reg))
        
    # create a temp subset
    lines_sel = random.choices(lines, k=10)
    all_lines += lines_sel
        
    with open(config_save_dir / ('03_config_limit_' + tmp_dataset_name + '.txt'), 'w') as f:
        for line in lines:
            f.write(line)
            f.write('\n')
            
    print(len(lines))
```

```python
all_lines
```

```python
with open(config_save_dir / ('03_config_limit_subset.txt'), 'w') as f:
        for line in all_lines:
            f.write(line)
            f.write('\n')
```
